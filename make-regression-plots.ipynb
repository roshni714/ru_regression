```python
import pandas as pd
import matplotlib.pyplot as plt
from data_loaders import generate_shift_one_dim_dataset
```

```python
X_train, y_train= generate_shift_one_dim_dataset(n=750, p=0.2, seed=0)
```

```python
plt.scatter(X_train.flatten(), y_train)
```

```python
standard_erm_df = pd.read_csv("results/inference_res_shifted_one_dim_erm_1_squared_loss_p_train_0.2_seed_0.csv")
oracle_erm_df = pd.read_csv("results/inference_res_shifted_one_dim_erm_1_squared_loss_p_train_0.5_seed_0.csv")
ru_gamma_16 = pd.read_csv("results/inference_res_shifted_one_dim_ru_regression_16.0_squared_loss_p_train_0.2_seed_0.csv")
ru_gamma_8 = pd.read_csv("results/inference_res_shifted_one_dim_ru_regression_8.0_squared_loss_p_train_0.2_seed_0.csv")
ru_gamma_4 = pd.read_csv("results/inference_res_shifted_one_dim_ru_regression_4.0_squared_loss_p_train_0.2_seed_0.csv")
ru_gamma_2 = pd.read_csv("results/inference_res_shifted_one_dim_ru_regression_2.0_squared_loss_p_train_0.2_seed_0.csv")
# ru_gamma_8 =  pd.read_csv("results/sim_shifted_ru_regression_8.0_squared_loss_seed_0.csv")
# ru_gamma_4 =  pd.read_csv("results/sim_shifted_ru_regression_4.0_squared_loss_seed_0.csv")
# ru_gamma_2 =  pd.read_csv("results/sim_shifted_ru_regression_2.0_squared_loss_seed_0.csv")
```

```python
plt.scatter(X_train.flatten(), y_train, alpha=0.2)
dfs = [standard_erm_df, ru_gamma_2, ru_gamma_4, ru_gamma_8, ru_gamma_16, oracle_erm_df]
names = ["Standard ERM",
         "RU-Reg " +r"$(\Gamma=2)$",
         "RU-Reg " +r"$(\Gamma=4)$",
         "RU-Reg " +r"$(\Gamma=8)$",
         "RU-Reg " +r"$(\Gamma=16)$",
         "Oracle ERM"
        ]

for i, df in enumerate(dfs):
    plt.plot(df["X"].to_numpy().flatten(), df["y"].to_numpy().flatten(), label=names[i])

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Learned Regression Functions (Superimposed on Training Data)")
```

```python
dfs = [standard_erm_df, ru_gamma_2, ru_gamma_4, ru_gamma_8, ru_gamma_16, oracle_erm_df]
names = ["Standard ERM",
         "RU-Reg " +r"$(\Gamma=2)$",
         "RU-Reg " +r"$(\Gamma=4)$",
         "RU-Reg " +r"$(\Gamma=8)$",
         "RU-Reg " +r"$(\Gamma=16)$",
         "Oracle ERM"
        ]

for i, df in enumerate(dfs):
    plt.plot(df["X"].to_numpy().flatten(), df["y"].to_numpy().flatten(), label=names[i])

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Learned Regression Functions")
```

```python
fig, ax =plt.subplots(1, 2, figsize=(15, 5))
ax[0].scatter(X_train.flatten(), y_train, alpha=0.2)
dfs = [standard_erm_df, ru_gamma_2, ru_gamma_4, ru_gamma_8, ru_gamma_16, oracle_erm_df]
names = ["Standard ERM",
         "RU Reg " +r"$(\Gamma=2)$",
         "RU Reg " +r"$(\Gamma=4)$",
         "RU Reg " +r"$(\Gamma=8)$",
         "RU Reg " +r"$(\Gamma=16)$",
         "Oracle ERM",
        ]

for i, df in enumerate(dfs):
    ax[0].plot(df["X"].to_numpy().flatten(), df["y"].to_numpy().flatten(), label=names[i])
    ax[1].plot(df["X"].to_numpy().flatten(), df["y"].to_numpy().flatten(), label=names[i])

for i in range(2):
    ax[i].set_xlabel(r"$x$",fontsize=15)
    ax[i].set_ylabel(r"$y$",fontsize=15)
    ax[i].legend()
ax[0].set_title("Learned Regression Functions\nData Distribution when "+ r"$p=0.2$", fontsize=20)
ax[1].set_title("Learned Regression Functions",fontsize=20)
plt.savefig("figs/learned_regression.pdf")
```
