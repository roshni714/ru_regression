```python
import pandas as pd
import matplotlib.pyplot as plt
from data_loaders import generate_shift_dataset
```

```python
X_train, y_train= generate_shift_dataset(n=500, p_d1=0.8)
```

```python
def standardize(data):
    mu = data.mean(axis=0, keepdims=1)
    scale = data.std(axis=0, keepdims=1)
    scale[scale < 1e-10] = 1.0
    data = (data - mu) / scale
    return data, mu, scale
```

```python
X_train, x_train_mu, x_train_scale = standardize(X_train)
y_train, y_train_mu, y_train_scale = standardize(y_train)
```

```python
plt.scatter(X_train.flatten(), y_train)
```

```python
standard_erm_df = pd.read_csv("results/sim2_shifted_erm_1_squared_loss_seed_3.csv")
oracle_erm_df = pd.read_csv("results/sim2_shifted_oracle_erm_1_squared_loss_seed_3.csv")
ru_gamma_16 = pd.read_csv("results/sim2_shifted_ru_regression_16.0_squared_loss_seed_3.csv")
ru_gamma_8 =  pd.read_csv("results/sim2_shifted_ru_regression_8.0_squared_loss_seed_3.csv")
ru_gamma_4 =  pd.read_csv("results/sim2_shifted_ru_regression_4.0_squared_loss_seed_3.csv")
ru_gamma_2 =  pd.read_csv("results/sim2_shifted_ru_regression_2.0_squared_loss_seed_3.csv")
```

```python
plt.scatter(X_train.flatten(), y_train, alpha=0.1)
dfs = [standard_erm_df, ru_gamma_2, ru_gamma_4, ru_gamma_8, ru_gamma_16, oracle_erm_df]
names = ["Standard ERM",
         "RU-Reg " +r"$(\Gamma=2)$",
         "RU-Reg " +r"$(\Gamma=4)$",
         "RU-Reg " +r"$(\Gamma=8)$",
         "RU-Reg " +r"$(\Gamma=16)$",
         "Oracle ERM"
        ]

for i, df in enumerate(dfs):
    plt.plot(df["X"].to_numpy().flatten(), df["y"].to_numpy().flatten(), label=names[i])

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```
